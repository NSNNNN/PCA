```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1) Generate clean signal
t = np.linspace(0, 1, 500)
signal = np.sin(2 * np.pi * 5 * t)

# 2) Add noise
noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + noise

# 3) Prepare data for PCA
X = np.vstack([noisy_signal, noisy_signal]).T

# 4) Apply PCA
pca = PCA(n_components=1)
X_denoised = pca.fit_transform(X)
X_reconstructed = pca.inverse_transform(X_denoised)

# 5) Plot results
plt.figure()
plt.plot(t, signal, label='Original Signal')
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.plot(t, X_reconstructed[:,0], label='After PCA')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noise Reduction using PCA')
plt.show()
```
